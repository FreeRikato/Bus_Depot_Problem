# **🚌 Bus Depot Problem**

**🧠 Algorithm:** Vogel's Approximation Method with a Total Cost Matrix

### **⚙️ Process**

1. **Initialization**: 
   * Set up supply (buses), demand (depots), and the cost matrix. 🚌 
2. **Penalty Calculation**:  
   * For each row/column, find the difference between the two lowest costs. This shows potential savings! 📉
3. **Allocation**: 
   * Pick the row/column with the biggest penalty.
   * Assign as much as you can to the cheapest spot in that row/column.
   * Update supply and demand.  ✅
4. **Iteration**: Repeat steps 2 and 3 until everything is assigned. 🔁 
5. **Total Cost Calculation**: Add up the costs of all the assignments. 💰

### **📦 Imported Packages**

* **pandas**: To work with data tables (DataFrames) 📊
* **numpy**: For number crunching  and arrays 🔢
* **random**: To add some randomness  to the simulation🎲
* **rich** (Maybe): For making the output look fancier ✨

### **📝 Variable Description**

* **no_of_depots (int)**: How many depots we have  🏭
* **depot_capacities (list)**: How much space each depot has 📦
* **df_demand (dictionary)**: How many buses each depot needs (key: depot name, value: demand)   {'D1': 20, 'D2': 35, ...} 🚌 
* **total_demand (int)**: Total buses needed across all depots 🚌🚌🚌...
* **df_supply (dictionary)**: Available buses (usually starts with 1 bus per entry) 🚌
* **total_supply (int)**:  Total number of available buses 🚌🚌🚌...
* **df_dkm (dataframe)**: Distances or costs to move an empty bus between locations 🛣️ 
* **vehicle_ages (list)**: How old each bus is 🚌👴
* **age (dictionary)**:  Maps each bus to its age 
* **kpl (list)**: Kilometers per liter (fuel efficiency) for each bus, based on age ⛽️
* **co2_emissions (list)**: CO2 emissions for each bus, based on age 💨
* **fc_kpl (dictionary)**: Fuel cost per kilometer for each bus 💸
* **co2_c (dictionary)**: Cost related to CO2 emissions for each bus 💸
* **df_doc (list)**: Depot operating costs  💼
* **df_tdkom (dataframe)**: The big one! Total cost to park a bus at each depot 💰

```python
def generate_TCM():
    # Set seed for reproducibility
    random.seed(1)

    # Initialize depots and their capacities
    # Number of Depots = 20 and Capacities = random integer between 100 and 150
    no_of_depots = 20
    depot_capacities = [random.randint(50, 150) for _ in range(no_of_depots)]
    df_demand = {'D{}'.format(i+1): cap for i, cap in enumerate(depot_capacities)}
    total_demand = sum(depot_capacities)
    
    # Calculate total supply

    # Generate supply data
    df_supply = {i: 1 for i in range(total_demand)}

    total_supply = sum(df_supply.values())
    
    np.random.seed(1)
    # Generate random dead kilometre matrix
    df_dkm = pd.DataFrame(np.random.uniform(0.5, 20, size=(total_supply, no_of_depots))).astype(float)
    
    random.seed(1)

    # Generate vehicle age
    vehicle_ages = [random.randint(1, 3) for _ in range(total_supply)]
    age = {i: age for i, age in enumerate(vehicle_ages)}

    # Generate kpl (kilometers per liter) and co2 emissions based on age
    random.seed(1) # Use the same random seed
    kpl, co2_emissions = [], []
    for v_age in vehicle_ages:
        if v_age == 1:
            kpl.append(random.uniform(5.1, 6))
            co2_emissions.append(round(random.uniform(515, 524), 2))
        elif v_age == 2:
            kpl.append(random.uniform(4.1, 5))
            co2_emissions.append(round(random.uniform(525, 534), 2))
        else:
            kpl.append(random.uniform(3.1, 4))
            co2_emissions.append(round(random.uniform(535, 540), 2))

    # Calculate fuel consumption, co2 cost and Depot Operating Cost
    fc_kpl = {i: (111 / kpl_val) for i, kpl_val in enumerate(kpl)}
    co2_c = {i: (0.00118642 * co2_val) for i, co2_val in enumerate(co2_emissions)}

    random.seed(1)
    df_doc = [random.uniform(50, 100) for _ in range(no_of_depots)]
    
    # Convert fc_kpl and co2_c dictionaries to pandas Series for efficient operations
    fc_kpl_series = pd.Series(fc_kpl)
    co2_c_series = pd.Series(co2_c)

    # Calculate the distance-based costs (fuel consumption and CO2 emissions) for each depot-destination pair
    distance_based_costs = df_dkm.multiply(fc_kpl_series, axis='index') + df_dkm.multiply(co2_c_series, axis='index')

    # Convert doc list to a pandas Series and align it with the columns of df_dkm for broadcasting
    doc_series = pd.Series(df_doc, index=df_dkm.columns)

    # Add the fixed depot operation costs to the distance-based costs
    # Broadcasting ensures that doc_series values are added to each row corresponding to the depot
    df_tdkom = distance_based_costs.add(doc_series, axis='columns')
    
    if input("Do you want to display the details of the total cost matrix? (Yes/No): ").lower() == "yes": display_TCM(no_of_depots, depot_capacities, total_demand, total_supply,df_supply, df_demand, df_dkm, kpl, co2_emissions, fc_kpl, co2_c, df_doc, df_tdkom)
    
    df_tdkom.columns = [("D"+str(i)) for i in range(1,21)]
    dk = df_tdkom.transpose()
    costs_t = dk.to_dict()
    cols = sorted(df_demand.keys())
    costs1=copy.deepcopy(costs_t)
    
    return costs_t, df_supply, df_demand, costs1, df_tdkom
```

```python
def SSM(costs,supply,demand):
    
    cost_f = copy.deepcopy(costs)
    df_sup =copy.deepcopy(supply)
    df_dem =copy.deepcopy(demand)

    res = dict((k, defaultdict(int)) for k in cost_f)
    g = {}
    for x in df_dem:
        g[x] = sorted(cost_f.keys(), key=lambda g: cost_f[g][x])

    for x in df_dem:
        res[g[x][0]][x] = df_dem[x]

    ## Deleting Rows if Supply is Met after First Allocation!

    #from operator import*
    for x in res:
        if list(res[x].values()) == []:
            #print('Null is:',x)
            continue
        else:
            #print(x)
            result = 0
            for i in list(res[x].values()):
                result += i
                #result = add(i, result)
            #print(result)
            if result == df_sup[x]:
                del df_sup[x]
                del cost_f[x]
        #print(x)

    while(df_sup):
        g = {}
        for x in df_dem:
            g[x] = sorted(cost_f.keys(), key=lambda g: cost_f[g][x])

        #### Step 3: To check the ER
        #from operator import*
        ER = []
        for x in df_sup:
            #### When only one cell is allocated a value in an ER
            if len(res[x].values()) == 1:
                if list(res[x].values())[0] > df_sup[x]:
                    ER.append(x)
                #print(ER)
            else:
                #### When multiple cells are allocated values in ER 
                result = 0
                for i in list(res[x].values()):
                    #result = add(i, result)
                    result += i
                if result > df_sup[x]:
                    ER.append(x)

                #print('Multiple cells:')
                #print(result,x)

        ## Step 4: To Calculate Differences for every cell in ER
        Diff = {}
        for x in ER:
            for y in res[x]:
                if len(g[y]) !=1:
                    if res[x][y] !=0:
                        m = 1
                        while g[y][m] in ER:
                            m += 1
                    #print(costs_t[g[y][0]][y], costs_t[g[y][m]][y])
                    Diff.setdefault(x, {}).setdefault(y, abs(cost_f[g[y][0]][y] - cost_f[g[y][m]][y]))

                elif len(g[y]) ==1:
                    Diff.setdefault(x, {}).setdefault(y, abs(cost_f[g[y][0]][y]))

        #print(Diff)

        SD = {key: min(val.values()) for key, val in Diff.items()}

        SDiff = min(SD.values())

        ### Counting occurrences of SDiff
        count = 0
        for key, value in SD.items():
            if value == SDiff:
                count += 1

        ### Step 5: For two/more SDiff with same value, choosing cell with largest allocation unit == (Will Always be SLC!)
        if count > 1:
            All = {}
            for k,v in Diff.items():
                #print(v)
                for (m,o) in v.items():
                    if o == SDiff:
                        All[k] = m

            max_val = 0
            for (key,value) in All.items():
                if res[key][value] > max_val:
                    max_val = res[key][value]
                    new_key = key
                    sub_key = value
                    #print(new_key,sub_key, res[new_key][sub_key])
                else:
                    continue

            new_value = SDiff
            #print(new_value, new_key, sub_key)

        if count == 1:
            new_value, new_key = min((value, key) for key, value in SD.items())

            key_list = list(Diff[new_key].keys())
            val_list = list(Diff[new_key].values())

            position = val_list.index(new_value)
            sub_key = key_list[position]

            #print(new_value, new_key, sub_key)

        ## If FLC Row cannot be satisfied & has more than one allocation:
        ## ~~ To ensure SLC must not be from an ER!
        if (len(Diff[new_key]) > 1):
            res_f = 0
            #for x in list(res[g[sub_key][0]].values()):
            for x in list(res[new_key].values()):
                res_f +=x

            ### Adding allocations in SLC Row
            x = 1
            while g[sub_key][x] in ER:
                x += 1

            ## Replacing 1's with 'x'
            #if res_f > df_supply[g[sub_key][0]]:
            if res_f > df_sup[new_key]:
                result = 0
                for x in list(res[g[sub_key][x]].values()):
                    result +=x

                ### If possible, satisfying the SLC Row
                if result < df_sup[g[sub_key][x]]:
                    m = df_sup[g[sub_key][x]] - result
                    res[g[sub_key][x]][sub_key] = m
                    res[new_key][sub_key] -= m
                    if res[new_key][sub_key] == 0:
                        del res[new_key][sub_key]
                    #res[g[sub_key][0]][sub_key] -= m
                    #if res[g[sub_key][0]][sub_key] == 0:
                    #    del res[g[sub_key][0]][sub_key]

                ### If it is not possible to satisfy the SLC Row, move all All.units from current FLC to SLC
                if result > df_sup[g[sub_key][x]]:
                    #m = res[g[sub_key][0]][sub_key]
                    m = res[new_key][sub_key]
                    res[g[sub_key][x]][sub_key] += m
                    del res[new_key][sub_key]
                    #del res[g[sub_key][0]][sub_key]

        ## If FLC Row can be satisfied & has only one allocation:
        if (len(Diff[new_key])==1): 

            ###~~~~~~~~~~~ CORRECTION !!!!!

            x = 1
            while g[sub_key][x] in ER:
                x += 1
            #print(x)
            if res[new_key][sub_key] == 1:
                res[g[sub_key][x]][sub_key] = res[new_key][sub_key]
                del res[new_key][sub_key]
            else:
                res[g[sub_key][x]][sub_key] = res[new_key][sub_key] - df_sup[new_key]
                res[new_key][sub_key] = df_sup[new_key]

        ### Adding up Cells in Rows
        result_1 = 0
        for i in list(res[new_key].values()):
            result_1 += i
        #print(result_1)

        result_2 = 0
        for i in list(res[g[sub_key][x]].values()):
            result_2 += i
        #print(result_2)

        if result_1 == df_sup[new_key]:
            del df_sup[new_key]
            del cost_f[new_key]

        if result_2 == df_sup[g[sub_key][x]]:
            del df_sup[g[sub_key][x]]
            del cost_f[g[sub_key][x]]

    cost = 0
    #costs1=copy.deepcopy(costs_t)
    cols = sorted(df_demand.keys())
    
    for g in sorted(costs1):
        # print (g, " ",)
        for n in cols:
            y = res[g][n]
            if y != 0:
                pass
                # print (y,)
            cost += y * costs1[g][n]
            # print ("  ",)
        # print(" ")
    print ("Total Cost yielded by SSM = ", cost)
```

```python
def BCE(costs,supply,demand):
    
    cost_f = copy.deepcopy(costs)
    df_sup =copy.deepcopy(supply)
    df_dem =copy.deepcopy(demand)
    
    res = dict((k, defaultdict(int)) for k in cost_f)
    g = {}
    for x in df_dem:
        g[x] = sorted(cost_f.keys(), key=lambda g: cost_f[g][x])

    for x in df_dem:
        res[g[x][0]][x] = df_dem[x]
        
    ## Deleting Rows if Supply is Met after First Allocation!

    # from operator import*
    for x in res:
        if list(res[x].values()) == []:
            #print('Null is:',x)
            continue
        else:
            #print(x)
            result = 0
            for i in list(res[x].values()):
                result += i
                #result = add(i, result)
            #print(result)
            if result == df_sup[x]:
                del df_sup[x]
                del cost_f[x]
        #print(x)
        
    while(df_sup):
        g = {}
        for x in df_dem:
            g[x] = sorted(cost_f.keys(), key=lambda g: cost_f[g][x])

        #### Step 3: To check the ER
        #from operator import*
        ER = []
        for x in df_sup:
            #### When only one cell is allocated a value in an ER
            if len(res[x].values()) == 1:
                if list(res[x].values())[0] > df_sup[x]:
                    ER.append(x)
                #print(ER)
            else:
                #### When multiple cells are allocated values in ER 
                result = 0
                for i in list(res[x].values()):
                    #result = add(i, result)
                    result += i
                if result > df_sup[x]:
                    ER.append(x)


        ## Step 4: To Calculate Differences for every cell in ER
        Diff = {}
        for x in ER:
            for y in res[x]:
                if len(g[y]) !=1:
                    if res[x][y] !=0:
                        m = 1
                        while g[y][m] in ER:
                            m += 1
                    #print(costs_t[g[y][m]][y], costs_t[g[y][0]][y])
                    Diff.setdefault(x, {}).setdefault(y, cost_f[g[y][m]][y] - cost_f[g[y][0]][y])         

                elif len(g[y]) ==1:
                    Diff.setdefault(x, {}).setdefault(y, cost_f[g[y][0]][y])

        #print(Diff)

        SD = {key: min(val.values()) for key, val in Diff.items()}

        SDiff = min(SD.values())

        ### Counting occurrences of SDiff
        count = 0
        for key, value in SD.items():
            if value == SDiff:
                count += 1

        ### Step 5: For two/more SDiff with same value, choosing cell with largest allocation unit!
        if count > 1:
            All = {}
            for k,v in Diff.items():
                #print(v)
                for (p,o) in v.items():
                    if o == SDiff:
                        All[k] = p

            max_val = 0
            for (key,value) in All.items():
                if res[key][value] > max_val:
                    max_val = res[key][value]
                    new_key = key
                    sub_key = value
                    #print(new_key,sub_key, res[new_key][sub_key])
                else:
                    continue

            new_value = SDiff
            #print(new_value, new_key, sub_key)

        if count == 1:
            new_value, new_key = min((value, key) for key, value in SD.items())

            key_list = list(Diff[new_key].keys())
            val_list = list(Diff[new_key].values())

            position = val_list.index(new_value)
            sub_key = key_list[position]

            #print(new_value, new_key, sub_key)

        FLR = new_key
        #FLR

        m = 1
        while g[sub_key][m] in ER:
            m += 1
        #print(m)
        SLR = g[sub_key][m]
        #SLR

        TCFLR = 0
        for i in list(cost_f[FLR].values()):
            TCFLR += i

        TCSLR = 0
        for i in list(cost_f[SLR].values()):
            TCSLR += i

        ### Step 9:
        if TCFLR > TCSLR:
            ## Satisfy FLR & keep the rest in SLC
            if res[FLR][sub_key] == 1:
                res[SLR][sub_key] = res[FLR][sub_key]
                del res[FLR][sub_key]
            else:
                res[SLR][sub_key] = res[FLR][sub_key] - df_sup[FLR]
                res[FLR][sub_key] = df_sup[FLR]


        ### Step 9:
        if TCFLR < TCSLR:
            ## Satisfy SLR & Keep the rest in FLC
            if res[FLR][sub_key] == 1:
                res[SLR][sub_key] = res[FLR][sub_key]
                del res[FLR][sub_key]
            else:
                res[FLR][sub_key] = res[FLR][sub_key] - df_sup[FLR]
                res[SLR][sub_key] = df_sup[SLR]


        result_1 = 0
        for i in list(res[FLR].values()):
            result_1 += i
        #print(result_1)

        result_2 = 0
        for i in list(res[SLR].values()):
            result_2 += i
        #print(result_2)

        if result_1 == df_sup[FLR]:
            del df_sup[FLR]
            del cost_f[FLR]

        if result_2 == df_sup[SLR]:
            del df_sup[SLR]
            del cost_f[SLR]
            
    cost = 0
    cols = sorted(df_demand.keys())
    for g in sorted(costs1):
        # print (g, " ",)
        for n in cols:
            y = res[g][n]
            if y != 0:
                pass
                # print (y,)
            cost += y * costs1[g][n]
            # print ("  ",)
        # print(" ")
    print ("Total Cost yielded by BCE = ", cost)
```

```python
def IVAM(costs_t, df_supply, df_demand):
    
    df_sup =copy.deepcopy(df_supply)
    df_dem =copy.deepcopy(df_demand)
    df_cost = df_tdkom.copy(deep=True)
    df_cost.columns = [i for i in range(0,20)]
    #print(df_cost)
    
    df_tdk = pd.DataFrame.from_dict(costs_t).transpose()
    df_tdk.columns = [i for i in range(0,20)]
    #print(df_tdkom)
    
    #df_cost = df_tdkom.copy(deep=True)
        
    translate = {}
    i = 0
    for k, v in df_dem.items():
        #print("This is the key: '%s' and this is the value '%s'\n" % (k, v) )
        new_key = i
        translate[k] = new_key
        #time.sleep(4)
        i+=1

    for old, new in translate.items():
        df_dem[new] = df_dem.pop(old)
    
    df_result = pd.DataFrame(np.zeros(df_tdk.shape))
    #print(df_result)
     

    while(list(df_sup.values()) and list(df_dem.values())):

        # To find the minimum cost row-wise
        df_nr_list = list(df_tdk.min(axis=1))
        df_nr = {}
        for i in range(len(df_tdk.index)):
            df_nr[df_tdk.index[i]] = df_nr_list[i]
        df_nr

        # To find the minimum cost column-wise
        df_nc_list = list(df_tdk.min(axis=0))
        df_nc = {}
        for i in range(len(df_tdk.columns)):
            df_nc[df_tdk.columns[i]] = df_nc_list[i]
        df_nc

        # To find the two lowest values row-wise and penalty cost
        df_pen = []
        if len(df_tdk.columns)!=1:
            df_pend = df_tdk.apply(lambda x: x.sort_values().unique(), axis=1)
            for j in df_tdk.index:
                if (len(df_pend[j])) != 1:
                    lst = df_pend[j][1] - df_pend[j][0]
                    df_pen.append(lst)
                elif (len(df_pend[j])) == 1: ### To check if there is same value across all rows
                    df_pen.append(0)
        else:
            df_pen = df_tdk.apply(lambda x: x.sort_values().unique()[0], axis=1)

        df_penr = {}
        if (df_tdk.shape[1] != 1):
            for i in range(len(df_tdk.index)):
                df_penr[df_tdk.index[i]] = df_pen[i]
        else:
            for i in df_tdk.index:
                df_penr[i] = df_pen[i]


        ################################################### *** ##############################################################
        # To find the two lowest values column-wise
        df_penc = []
        if len(df_tdk.index)!=1:

            df_pendo = df_tdk.apply(lambda x: x.sort_values().unique(), axis=0)

            if (type(df_pendo) == pd.DataFrame):
                for i in df_tdk.columns:
                    if (len(df_pendo[i])!=1):
                        lstc = df_pendo[i].values[1] - df_pendo[i].values[0]
                        df_penc.append(lstc)
                    elif (len(df_pendo[i])==1):
                        df_penc.append(0)
            elif (type(df_pendo) == pd.Series):
                for j in df_tdk.columns:
                    if (len(df_pendo[j])) != 1:
                        lstc = df_pendo[j][1] - df_pendo[j][0]
                        df_penc.append(lstc)
                    elif (len(df_pendo[j])) == 1: ### To check if there is same value across all rows
                        df_penc.append(0)

        else:
            df_penc = df_tdk.apply(lambda x: x.sort_values().unique()[0], axis=0)

        df_penco = {}
        if (df_tdk.shape[0] != 1):
            for i in range(len(df_tdk.columns)):
                df_penco[df_tdk.columns[i]] = df_penc[i]
        else:
            for i in df_tdk.columns:
                df_penco[i] = df_penc[i]

        r = max(list(df_penr.values()) + list(df_penco.values()))

        lstm = []
        lst = []
        lstr = []
        if(r in df_penco.values() and r in df_penr.values()):
            min_val = []
            idx_list = []
            for i in df_tdk.index:
                for j in (df_tdk.columns):
                    if (df_penr[i] == r and df_penco[j] == r):
                        if (df_nr[i] <= df_nc[j]):
                            for l in df_tdk.columns:
                                if (df_tdk.loc[i,l] == df_nr[i]):
                                    #print(df_tdkom.loc[i,l], df_nr[i])
                                    min_val.append(df_nr[i])
                                    idx_list.append([i,l])
                                    #lstm.append([i,l])
                        elif (df_nr[i] > df_nc[j]):
                            for o in df_tdk.index:
                                 if (df_tdk.loc[o,j] == df_nc[j]):
                                        min_val.append(df_nc[j])
                                        idx_list.append([o,j])
                                        #lstm.append([o,j])

                        lstm.append(idx_list[min_val.index(min(min_val))])


            ###### Allocating to the left most cell, since that' how it identifies
            df_result.loc[lstm[0][0],lstm[0][1]] = min(df_sup[lstm[0][0]],df_dem[lstm[0][1]])
            df_sup[lstm[0][0]] = df_sup[lstm[0][0]] - df_result.loc[lstm[0][0],lstm[0][1]]
            df_dem[lstm[0][1]] = df_dem[lstm[0][1]] - df_result.loc[lstm[0][0],lstm[0][1]]


            if (df_sup[lstm[0][0]] == 0 and df_dem[lstm[0][1]] == 0):
                #print(df_supply[lstm[0][0]], df_demand[lstm[0][1]] )
                df_tdk = df_tdk.drop(lstm[0][0], axis = 0)
                del df_sup[lstm[0][0]]

            elif (df_sup[lstm[0][0]] == 0 and df_dem[lstm[0][1]] != 0):
                #print(df_supply[lstm[0][0]], df_demand[lstm[0][1]] )
                df_tdk = df_tdk.drop(lstm[0][0], axis = 0)
                del df_sup[lstm[0][0]]

            elif (df_sup[lstm[0][0]] != 0 and df_dem[lstm[0][1]] == 0):
                #print(df_supply[lstm[0][0]], df_demand[lstm[0][1]] )
                df_tdk = df_tdk.drop(lstm[0][1], axis = 1)
                del df_dem[lstm[0][1]]


        elif(r in df_penco.values() and r not in df_penr.values()):
            for i in df_tdk.columns:
                if (df_penco[i] == r):
                    for j in (df_tdk.index):
                        if (df_tdk.loc[j,i] == df_nc[i]):
                            lst.append([j,i])

            if (len(lst) == 1):
                ###### Allocating to the left most cell, since that' how it identifies
                df_result.loc[lst[0][0],lst[0][1]] = min(df_sup[lst[0][0]],df_dem[lst[0][1]])
                df_sup[lst[0][0]] = df_sup[lst[0][0]] - df_result.loc[lst[0][0],lst[0][1]]
                df_dem[lst[0][1]] = df_dem[lst[0][1]] - df_result.loc[lst[0][0],lst[0][1]]


                if (df_sup[lst[0][0]] == 0 and df_dem[lst[0][1]] == 0):
                    #print(df_supply[lst[0][0]], df_demand[lst[0][1]] )
                    df_tdk = df_tdk.drop(lst[0][0], axis = 0)
                    del df_sup[lst[0][0]]

                elif (df_sup[lst[0][0]] == 0 and df_dem[lst[0][1]] != 0):
                    #print(df_supply[lst[0][0]], df_demand[lst[0][1]] )
                    df_tdk = df_tdk.drop(lst[0][0], axis = 0)
                    del df_sup[lst[0][0]]

                elif (df_sup[lst[0][0]] != 0 and df_dem[lst[0][1]] == 0):
                    #print(df_supply[lst[0][0]], df_demand[lst[0][1]] )
                    df_tdk = df_tdk.drop(lst[0][1], axis = 1)
                    del df_dem[lst[0][1]]

            elif (len(lst)!=1):
                minc = df_tdk.loc[lst[0][0],lst[0][1]]
                for i in range(len(lst)):
                    if (minc <= df_tdk.loc[lst[i][0],lst[i][1]]):
                        minc = minc
                    elif (minc > df_tdk.loc[lst[i][0],lst[i][1]]):
                        minc = df_tdk.loc[lst[i][0],lst[i][1]]

                lstn = []
                for i in range(len(lst)):
                    if (df_tdk.loc[lst[i][0],lst[i][1]] == minc):
                        lstn.append([lst[i][0],lst[i][1]])

                df_result.loc[lstn[0][0],lstn[0][1]] = min(df_sup[lstn[0][0]],df_dem[lstn[0][1]])
                df_sup[lstn[0][0]] = df_sup[lstn[0][0]] - df_result.loc[lstn[0][0],lstn[0][1]]
                df_dem[lstn[0][1]] = df_dem[lstn[0][1]] - df_result.loc[lstn[0][0],lstn[0][1]]

                if (df_sup[lstn[0][0]] == 0 and df_dem[lstn[0][1]] == 0):
                    df_tdk = df_tdk.drop(lstn[0][0], axis = 0)
                    del df_sup[lstn[0][0]]

                elif (df_sup[lstn[0][0]] == 0 and df_dem[lstn[0][1]] != 0):
                    df_tdk = df_tdk.drop(lstn[0][0], axis = 0)
                    del df_sup[lstn[0][0]]

                elif (df_sup[lstn[0][0]] != 0 and df_dem[lstn[0][1]] == 0):
                    df_tdk = df_tdk.drop(lstn[0][1], axis = 1)
                    del df_dem[lstn[0][1]]



        elif (r in df_penr.values() and r not in df_penco.values()):
            for i in df_tdk.index:
                if (df_penr[i] == r):
                    for j in (df_tdk.columns):
                        if (df_tdk.loc[i,j] == df_nr[i]):
                            lstr.append([i,j])

            if (len(lstr) == 1):
                #print(lstr)
                df_result.loc[lstr[0][0],lstr[0][1]] = min(df_sup[lstr[0][0]],df_dem[lstr[0][1]])
                df_sup[lstr[0][0]] = df_sup[lstr[0][0]] - df_result.loc[lstr[0][0],lstr[0][1]]
                df_dem[lstr[0][1]] = df_dem[lstr[0][1]] - df_result.loc[lstr[0][0],lstr[0][1]]

                if (df_sup[lstr[0][0]] == 0 and df_dem[lstr[0][1]] == 0):
                    df_tdk = df_tdk.drop(lstr[0][0], axis = 0)
                    del df_sup[lstr[0][0]]

                elif (df_sup[lstr[0][0]] == 0 and df_dem[lstr[0][1]] != 0):
                    df_tdk = df_tdk.drop(lstr[0][0], axis = 0)
                    del df_sup[lstr[0][0]]

                elif (df_sup[lstr[0][0]] != 0 and df_dem[lstr[0][1]] == 0):
                    df_tdk = df_tdk.drop(lstr[0][1], axis = 1)
                    del df_dem[lstr[0][1]]

            elif (len(lstr)!=1):
                minr = df_tdk.loc[lstr[0][0],lstr[0][1]]
                for i in range(len(lstr)):
                    if (minr <= df_tdk.loc[lstr[i][0],lstr[i][1]]):
                        minr = minr
                    elif (minr > df_tdk.loc[lstr[i][0],lstr[i][1]]):
                        minr = df_tdk.loc[lstr[i][0],lstr[i][1]]

                lstp = []
                for i in range(len(lstr)):
                    if (df_tdk.loc[lstr[i][0],lstr[i][1]] == minr):
                        lstp.append([lstr[i][0],lstr[i][1]])

                df_result.loc[lstp[0][0],lstp[0][1]] = min(df_sup[lstp[0][0]],df_dem[lstp[0][1]])
                df_sup[lstp[0][0]] = df_sup[lstp[0][0]] - df_result.loc[lstp[0][0],lstp[0][1]]
                df_dem[lstp[0][1]] = df_dem[lstp[0][1]] - df_result.loc[lstp[0][0],lstp[0][1]]

                if (df_sup[lstp[0][0]] == 0 and df_dem[lstp[0][1]] == 0):
                    df_tdk = df_tdk.drop(lstp[0][0], axis = 0)
                    del df_sup[lstp[0][0]]

                elif (df_sup[lstp[0][0]] == 0 and df_dem[lstp[0][1]] != 0):
                    df_tdk = df_tdk.drop(lstp[0][0], axis = 0)
                    del df_sup[lstp[0][0]]

                elif (df_sup[lstp[0][0]] != 0 and df_dem[lstp[0][1]] == 0):
                    df_tdk = df_tdk.drop(lstp[0][1], axis = 1)
                    del df_dem[lstp[0][1]]

    #print(df_result)
    
    df_final = df_cost.mul(df_result, axis = 0)    
    print ("Total Cost yielded by IVAM = ", df_final.to_numpy().sum())
   
```

```python
def RA(costs_t, df_supply, df_demand):
    
    df_sup =copy.deepcopy(df_supply)
    df_dem =copy.deepcopy(df_demand)
    df_cost = df_tdkom.copy(deep=True)
    df_cost.columns = [i for i in range(0,20)]
    #print(df_cost)
    
    df_tdk = pd.DataFrame.from_dict(costs_t).transpose()
    df_tdk.columns = [i for i in range(0,20)]
    #print(df_tdk)
    
    #df_cost = df_tdkom.copy(deep=True)
        
    translate = {}
    i = 0
    for k, v in df_dem.items():
        #print("This is the key: '%s' and this is the value '%s'\n" % (k, v) )
        new_key = i
        translate[k] = new_key
        #time.sleep(4)
        i+=1

    for old, new in translate.items():
        df_dem[new] = df_dem.pop(old)
    
    df_result = pd.DataFrame(np.zeros(df_tdk.shape))
    
    #df_cost = df_tdkom.copy(deep=True)
    
    # Ranking each depot depending on value of DK
    df_rank= (df_tdk.rank(1, ascending=True, method='first')).astype(int) 
    #print(df_rank)

    df_depot = df_dem.copy()

    R = 1
    
    while(R!=len(df_depot)+1): 
        df_rankr = (df_rank == R)*1.0

        lst = []
        for m in df_rankr.columns:
            if (df_rankr.sum(axis=0)[m] <= df_dem[m]):
                for j in df_rankr.index:
                    df_result.loc[j,m] = df_rankr.loc[j,m] 

        for m in df_rankr.columns:     
            if (df_rankr.sum(axis=0)[m] > df_dem[m]):
                    for j in df_rankr.index:
                        if (df_rank.loc[j,m] == R):
                            lst.append((j,m))

                    new = []
                    for (k,v) in lst:
                        if (v == m):
                            new.append(df_tdk.loc[k,v])      

                    new.sort()
                    n = int(df_dem[m])
                    #print(n)
                    newn = new[0:n]

                    for j in df_rankr.index:
                        for i in range(len(newn)):
                            if (newn[i] == df_tdk.loc[j,m]): 
                                df_result.loc[j,m] = df_rankr.loc[j,m]  

        for m in df_rankr.columns:
            df_dem[m] = df_depot[m] - df_result.sum(axis=0)[m] 

        for m in df_rankr.columns: 
            if (df_dem[m] == 0):
                    df_rank = df_rank.drop(m, axis=1)
                    del df_dem[m]

        for k in df_rankr.index:
            df_sup[k] = df_sup[k] - df_result.sum(axis=1)[k]

        for i in df_rankr.index:
            if (df_sup[i] == 0):
                df_rank = df_rank.drop(i, axis = 0)
                del df_sup[i]

        R = R + 1
    
    df_final = df_cost.mul(df_result, axis = 0)    
    print ("Total Cost yielded by RA = ", df_final.to_numpy().sum())
```

```python
def MDM(costs_t, df_supply, df_demand):
    
    df_sup =copy.deepcopy(df_supply)
    df_dem =copy.deepcopy(df_demand)
    df_cost = df_tdkom.copy(deep=True)
    df_cost.columns = [i for i in range(0,20)]
    #print(df_cost)
    
    df_tdk = pd.DataFrame.from_dict(costs_t).transpose()
    df_tdk.columns = [i for i in range(0,20)]
    #print(df_tdk)
    
    #df_cost = df_tdkom.copy(deep=True)
        
    translate = {}
    i = 0
    for k, v in df_dem.items():
        #print("This is the key: '%s' and this is the value '%s'\n" % (k, v) )
        new_key = i
        translate[k] = new_key
        #time.sleep(4)
        i+=1

    for old, new in translate.items():
        df_dem[new] = df_dem.pop(old)
    
    df_result = pd.DataFrame(np.zeros(df_tdk.shape))
    
    while(list(df_sup.values()) and list(df_dem.values())): 
        # To find the two lowest values column-wise
        if len(df_tdk.index)!=1:
            import heapq
            smallest = df_tdk.apply(lambda x: heapq.nsmallest(2, x), axis=0)
            smallest_df = pd.DataFrame(smallest.values.tolist())
            dfn = smallest_df.loc[1] - smallest_df.loc[0]
            df_pen = list(dfn)
        else:
            df_pen = list(df_tdk.apply(lambda x: x.sort_values().unique()[0], axis=0))

        df_penc = {}
        for i in range(len(df_tdk.columns)):
            df_penc[df_tdk.columns[i]] = df_pen[i]


        max_d = 0
        for (k,v) in df_dem.items():
            if (v > max_d):
                max_d = v

        kmx = [key for key,val in df_dem.items() if val == max_d] #keys with maximum demand value


        # To find the column with maximum penalty cost, or least cost cell (max_key = that column) 
        max = 0
        for i in range(len(kmx)):
            if (df_penc[kmx[i]] > max):
                max = df_penc[kmx[i]]


        kmpen = [key for key,val in df_penc.items() if val == max] #keys with same penalty cost


        # To find the cell with smallest unit cost in case of a VAM tie
        minval = []
        for i in kmpen:
            minval.append(df_tdk[i].min())
        p = min(minval)


        for i in kmpen:
            for j in df_tdk.index:
                if (p == df_tdk.loc[j,i]):
                    max_key = i #To store the columns in max_key with the least cost

        lst = []
        for j in df_tdk.columns:
            if (j == max_key): # For same values of demand, with different penalty cost
                #print(j)
                while (df_dem[j] != 0):
                    df_nc_ = df_tdk.min(axis=0)[j]
                    #print(df_nc_)
                    for i in df_tdk.index:
                        if(df_tdk.loc[i,j] == df_nc_):
                            lst.append([i,j])     
                            df_result.loc[lst[0][0],lst[0][1]] = min(df_sup[lst[0][0]],df_dem[lst[0][1]])
                            df_sup[lst[0][0]] = df_sup[lst[0][0]] - df_result.loc[lst[0][0],lst[0][1]]
                            df_dem[lst[0][1]] = df_dem[lst[0][1]] - df_result.loc[lst[0][0],lst[0][1]]
                            #print(df_result)
                            lst = []

                        if (df_sup[i] == 0):
                            df_tdk = df_tdk.drop(i, axis = 0)
                            del df_sup[i]

                if (df_dem[j] == 0):
                    df_tdk = df_tdk.drop(j, axis=1)
                    del df_dem[j]
                    
    df_final = df_cost.mul(df_result, axis = 0)    
    print ("Total Cost yielded by MDM = ", df_final.to_numpy().sum())
```
